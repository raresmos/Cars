# Average cylindrical capacity depending on the car year

```python
average_cmc = data.groupby('model_year')['displacement'].mean().round(2)
```

```python
average_cmc
```

```python
plt.plot(average_cmc.index , average_cmc.values)
plt.xlabel('Cars Year')
plt.ylabel('CMC')
plt.show  
```
